## Exploring the breast cancer dataset

This dataset contains 569 rows and 32 columns. These columns are measurements of properties of the tumor such as radius and perimeter.

```python
# Import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import warnings
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split, KFold, cross_val_score, RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

## Checking the data

The dataset has no missing values, and the datatype for each feature is correct.

```python
# Import the data
df = pd.read_csv('breast-cancer.csv')
```

```python
# Check the properties of each column
df.info()
```

```python
# Check the statistical summary of the columns
df.describe()
```

```python
# Check for missing values
df.isnull().sum()
```

## Exploratory Data Analysis

The benign cases are more than malignant cases.

```python
sns.countplot(x='diagnosis', data=df)
plt.title("Count of malignant vs benign cases")
plt.show()
```

```python
# Encode the diagnosis feature to a numerical variable
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

```python
# Check the correlation of each feature
df.corr()
```

```python
# Separate the target and feature variables
X = df.drop(columns=['id', 'diagnosis'])
y = df['diagnosis']
```

The dataset contains a lot of columns which have different measurement and scales. The StandardScaler is used to convert the dataframe into one with zero mean and unit variance (Standard Normal Distribution). The features have a high dimensionality and are reduced using PCA.

```python
# Scale the dataframe
sc = StandardScaler()
scaled_X = sc.fit_transform(X)
```

Scaling the dataframe converts it to an array. This array is then converted to a dataframe.

```python
# Convert the scaled dataframe to a dataframe
df_X_scaled = pd.DataFrame(scaled_X, columns=X.columns)
```

Using PCA to reduce the number of features gives an attribute called 'explained_variance_ratio_'. This is used to determine the best number of components/features to reduce to.

```python
arr = np.arange(31)
var_ratio = []
for i in arr:
    pca = PCA(n_components=i)
    pca.fit(df_X_scaled)
    var_ratio.append(np.sum(pca.explained_variance_ratio_))
```

The graph of the sum of the explained_variance_ratio for each number of component shows that when more than 7 components are used, over 90% of the variance can be explained by the new number of components. This covers a lot of the variance in the data.

```python
sns.set_style('darkgrid')
sns.lineplot(x=arr, y=var_ratio)
sns.scatterplot(x=arr, y=var_ratio)
plt.xlabel('n_components')
plt.ylabel('Explained variance ratio')
plt.title('Explained Variance Ratio vs n_components')
plt.show()
```

```python
# Reduce the number of compoents to 8
pca = PCA(n_components=8)
X_pca = pca.fit_transform(df_X_scaled)
```

```python
df_X_pca = pd.DataFrame(X_pca)
```

## Model selection and evaluation

The dataframe is split into train and test datasets. Four classification algorithms are fitted against the train data to determine which algorithm best fits the data by comparing the cross validation score of the algorithms.

```python
X_train, X_test, y_train, y_test = train_test_split(df_X_pca, y, random_state=42, test_size=0.25)
```

```python
results = []
models = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest Classifier': RandomForestClassifier(),
    'Decision Tree Classifier': DecisionTreeClassifier(),
    'K Nearest Neighbors': KNeighborsClassifier()
}
for model in models.values():
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    cv_scores = cross_val_score(model, X_train, y_train, cv=kf)
    results.append(cv_scores)
```

```python
plt.boxplot(results, labels=models.keys())
plt.xticks(rotation=45)
plt.title('Cross validation scores of models')
plt.xlabel('Model')
plt.ylabel('Cross validation score')
plt.show()
```

Logistic regression has the highest median cross validation score. Hence, it is used for prediction. RandomizedSearch is used to obtain the best hyperparameters for the model. On tuning, it has an accuracy score of 99.3%.

```python
# Set the grid parameters
param_grid = {
    'dual': [True, False],
    'max_iter': [100, 110, 120, 130, 140],
    'C': [1.0, 1.5, 2.0, 2.5]
}
```

```python
# Search for the best hyperparameters
warnings.filterwarnings('ignore')

lr = LogisticRegression(penalty='l2')
random = RandomizedSearchCV(estimator=lr, param_distributions=param_grid, cv=3, n_jobs=-1)
start_time = time.time()
random_result = random.fit(X_train, y_train)

print('Best score: {} using {}'.format(random_result.best_score_, random_result.best_params_))
print('Execution time: ' + str(time.time() - start_time) + ' ms')
```

```python
lr = LogisticRegression(C=2.0, max_iter=120)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

The confusion matrix gives the accuracy of prediction, where there are **0** false positives and **1** false negative. **89** cases are correctly predicted as `Benign` and **53** cases are correctly predicted as `Malignant`.

```python
ax = plt.subplot()
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, ax=ax)
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels')
ax.xaxis.set_ticklabels(['Benign', 'Malignant'])
ax.yaxis.set_ticklabels(['Benign', 'Malignant'])
plt.show()
```
